6- Explain the concept of generators in Python and how they are defined.

- generators are special types of functions that use iterators and provides the sequence of results lazily, every time "next" method is executed it provides result till it encounters next Yield in the generator, in this way way generator dont have to provide all result in one moment which consumes more memory.
- "yield" keyword is used instead of return to define a generator.

````python
def squareUpto(num):    # provides sequence of squares upto number num
  for i in range(1,num+1):
    yield i**2      # yield keyword used

gen= squareUpto(4)
print(type(gen))

print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
````

7- What are the advantages of using generators over regular functions?

- generators are special types of functions, that provides sequence of result lazily, unlike regular functions where result is generated all at once.
- in this way, regular functions consumes more memory when conpared to generators
- for large scale evaluation, generators are faster, as they consume less memory and less space


````python
# regular function
def regfunc(num):
  result=[]
  for i in range(1,num+1):
    result.append(i**2)
  return result

result= regfunc(4)
print(result)

# generator function
def genfunc(num):
  for i in range(1,num+1):
    yield i**2

gen= genfunc(4)
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
````

10- What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?



*   map()- this function maps every member of iterable to a particular function to get result corresponding to every member.
*   reduce()- this function reduces a sequence of iterable members to a single result with the help of function that takes 2 parameter and single resultant.
*   filter()- this function filters every member of iterable by following a conditional function that takes every member and results in true or false, based on which member is filtered out.



````python
lst1= [2,3,6,7,12]

# map
res1= list(map(lambda x:x**2, lst1))  # gives square of each member
print(res1)

from functools import reduce
# reduce
res2= reduce(lambda x,y: x if x>y else y, lst1)  # gives largest member of list
print(res2)

# filter
res3 = list(filter(lambda x: True if x%2==0 else False, lst1))  # gives even numbers from the list
print(res3)
````

1-  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

````python
def sumEven(numlst):
  sum=0
  for i in numlst:
    if i%2==0:
      sum+=i
  return sum

lst= [1,2,3,4,5,6]
print(sumEven(lst))
````

4- Write a Python function that checks if a given number is prime or not from 1 to 200

````python
def checkPrime(num):
  if num==1:
    return False
  for i in range(2, int(num/2)+1):
    if num % i==0:
      return False
  return True

print(checkPrime(11))
````

7-  Implement a generator function that reads a file line by line and yields each line as a string

image.png

````python
def line_generator(filename):
  with open(filename, mode='r') as file:
    for line in file:
      yield line.rstrip('\n')

gen= line_generator('python3test.txt')

print(next(gen))
print(next(gen))
print(next(gen))
````
